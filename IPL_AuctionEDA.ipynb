```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("D:\cricket_bid\Datasets\IPL_Auction_Data_2008_2025.csv")
```

```python
sns.set_style('whitegrid')
```

```python
# Distribution of Total Expenses
plt.figure(figsize=(10, 5))
sns.histplot(df["Total Expenses (Cr)"], bins=20, kde=True, color="blue")
plt.title("Distribution of Total Expenses (Cr)")
plt.xlabel("Total Expenses (Cr)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Distribution of Players Bought
plt.figure(figsize=(10, 5))
sns.histplot(df["Players Bought"], bins=15, kde=True, color="green")
plt.title("Distribution of Players Bought")
plt.xlabel("Players Bought")
plt.ylabel("Frequency")
plt.show()
```

```python
# Average spending by team over the years
plt.figure(figsize=(12, 6))
team_avg_spending = df.groupby("Team")["Total Expenses (Cr)"].mean().sort_values()
sns.barplot(x=team_avg_spending, y=team_avg_spending.index, palette="viridis")
plt.title("Average Total Expenses by Team (Cr)")
plt.xlabel("Average Total Expenses (Cr)")
plt.ylabel("Team")
plt.show()
```

```python
# Spending trends over the years
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Year", y="Total Expenses (Cr)", hue="Team", marker="o", palette="tab10")
plt.title("Total Expenses by Team Over the Years")
plt.xlabel("Year")
plt.ylabel("Total Expenses (Cr)")
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.show()
```

```python
# Remaining Purse Trends over the years
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Year", y="Remaining Purse (Cr)", hue="Team", marker="o", palette="tab10")
plt.title("Remaining Purse by Team Over the Years")
plt.xlabel("Year")
plt.ylabel("Remaining Purse (Cr)")
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x="Players Bought", y="Total Expenses (Cr)", hue="Team", palette="tab10", alpha=0.8)
plt.title("Players Bought vs Total Expenses")
plt.xlabel("Players Bought")
plt.ylabel("Total Expenses (Cr)")
plt.show()
```

```python
# Players Bought over the years
plt.figure(figsize=(12, 6))
players_bought_per_year = df.groupby("Year")["Players Bought"].sum()
sns.lineplot(x=players_bought_per_year.index, y=players_bought_per_year.values, marker="o", color="green")
plt.title("Total Players Bought Over the Years")
plt.xlabel("Year")
plt.ylabel("Players Bought")
plt.grid(True)
plt.show()
```

```python
# Top spending teams per year
top_spending_teams = df.loc[df.groupby("Year")["Total Expenses (Cr)"].idxmax()][["Year", "Team", "Total Expenses (Cr)"]]

plt.figure(figsize=(12, 6))
sns.barplot(data=top_spending_teams, x="Year", y="Total Expenses (Cr)", hue="Team", dodge=False, palette="magma")
plt.title("Top Spending Team Per Year")
plt.xlabel("Year")
plt.ylabel("Total Expenses (Cr)")
plt.xticks(rotation=45)
plt.legend(title="Team", loc="upper left", bbox_to_anchor=(1, 1))
plt.show()
```

```python
# Year-wise remaining purse trend
plt.figure(figsize=(12, 6))
yearly_purse = df.groupby("Year")["Remaining Purse (Cr)"].sum()
sns.lineplot(x=yearly_purse.index, y=yearly_purse.values, marker="o", color="blue")
plt.title("Remaining Purse Across All Teams Over the Years")
plt.xlabel("Year")
plt.ylabel("Remaining Purse (Cr)")
plt.grid(True)
plt.show()
```

```python
# Year-wise total spending trend
plt.figure(figsize=(12, 6))
yearly_spending = df.groupby("Year")["Total Expenses (Cr)"].sum()
sns.lineplot(x=yearly_spending.index, y=yearly_spending.values, marker="o", color="red")
plt.title("Total Spending Across All Teams Over the Years")
plt.xlabel("Year")
plt.ylabel("Total Expenses (Cr)")
plt.grid(True)
plt.show()
```
