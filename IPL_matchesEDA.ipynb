```python
import seaborn as sns 
```

```python
import matplotlib.pyplot as plt
```

```python
# Plot 1: Number of Matches Per Season
plt.figure(figsize=(10, 5))
sns.countplot(x=df['Date'].dt.year, palette='viridis')
plt.title('Number of Matches Per Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot 2: Top 10 Venues by Match Count
plt.figure(figsize=(12, 6))
top_venues = df['Venue'].value_counts().head(10)
sns.barplot(x=top_venues.values, y=top_venues.index, palette='coolwarm')
plt.title('Top 10 Venues by Number of Matches')
plt.xlabel('Number of Matches')
plt.ylabel('Venue')
plt.show()
```

```python
df.head(1)
```

```python
# Plot 3: Number of Wins by Team (Based on Winner Column)
plt.figure(figsize=(12, 6))
top_winners = df['Match_Winner'].value_counts().head(10)
sns.barplot(x=top_winners.values, y=top_winners.index, palette='magma')
plt.title('Top 10 Teams by Number of Wins')
plt.xlabel('Number of Wins')
plt.ylabel('Teams')
plt.show()
```

```python
# Plot 4: Toss Decision Analysis
plt.figure(figsize=(8, 5))
sns.countplot(x='Toss_Decision', data=df, palette='Set2')
plt.title('Toss Decisions (Fielding vs Batting)')
plt.xlabel('Toss Decision')
plt.ylabel('Count')
plt.show()
```

```python
# Plot 5: Player of the Match Analysis - Top 10 Players with Most Awards
plt.figure(figsize=(12, 6))
top_players = df['Player_of_Match'].value_counts().head(10)
sns.barplot(x=top_players.values, y=top_players.index, palette='plasma')
plt.title('Top 10 Players with Most Player of the Match Awards')
plt.xlabel('Number of Awards')
plt.ylabel('Players')
plt.show()
```

```python
df.head(1)
```

```python
# Plot 2: Win by Runs vs Win by Wickets Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['Win_Margin'][df['Win_Type'] == 'runs'], color='blue', label='Win by Runs', kde=True, bins=20)
sns.histplot(df['Win_Margin'][df['Win_Type'] == 'wickets'], color='red', label='Win by Wickets', kde=True, bins=10)
plt.title('Distribution of Wins by Runs and Wickets')
plt.xlabel('Margin')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Plot 7: Umpire Analysis - Top 10 Umpires
plt.figure(figsize=(12, 6))
umpire_list = pd.concat([df['Umpire1'], df['Umpire2']])
top_umpires = umpire_list.value_counts().head(10)
sns.barplot(x=top_umpires.values, y=top_umpires.index, palette='viridis')
plt.title('Top 10 Umpires by Number of Matches')
plt.xlabel('Number of Matches')
plt.ylabel('Umpires')
plt.show()
```

```python
# Create a pivot table for Head-to-Head match wins
head_to_head = df.pivot_table(index='Match_Winner', columns='Opponent', values='ID', aggfunc='count', fill_value=0)

# Sort teams based on total wins
head_to_head['Total Wins'] = head_to_head.sum(axis=1)
head_to_head = head_to_head.sort_values(by='Total Wins', ascending=False)

# Display top 10 teams
print(head_to_head.head(10))
```
