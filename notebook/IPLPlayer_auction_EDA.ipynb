```python
# Set style for better visuals
sns.set_style("whitegrid")
```

```python
# Set style for better visuals
sns.set_style("whitegrid")

# Plot distribution of Amount and Year
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Distribution of Amount (Auction Price)
sns.histplot(df["Amount"], bins=30, kde=True, ax=axes[0], color="blue")
axes[0].set_title("Distribution of Auction Amount")
axes[0].set_xlabel("Amount (INR)")
axes[0].set_ylabel("Frequency")
```

```python
# Distribution of Year
sns.countplot(x=df["Year"], ax=axes[1], palette="coolwarm")
axes[1].set_title("Number of Players Sold per Year")
axes[1].set_xlabel("Year")
axes[1].set_ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
# Player Roles Distribution
sns.countplot(y=df["Role"], ax=axes[0], palette="viridis", order=df["Role"].value_counts().index)
axes[0].set_title("Distribution of Player Roles")
axes[0].set_xlabel("Count")
axes[0].set_ylabel("Role")
```

```python
# Distribution of Auction Amount
plt.figure(figsize=(8, 5))
sns.histplot(df["Amount"], bins=30, kde=True, color="blue")
plt.title("Distribution of Auction Amount")
plt.xlabel("Amount (INR)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Number of Players Sold per Year
plt.figure(figsize=(10, 5))
sns.countplot(x=df["Year"], palette="coolwarm")
plt.title("Number of Players Sold per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Player Roles Distribution
plt.figure(figsize=(8, 5))
sns.countplot(y=df["Role"], palette="viridis", order=df["Role"].value_counts().index)
plt.title("Distribution of Player Roles")
plt.xlabel("Count")
plt.ylabel("Role")
plt.show()
```

```python
# Player Origin Distribution
plt.figure(figsize=(6, 5))
sns.countplot(x=df["Player Origin"], palette="muted")
plt.title("Indian vs Overseas Players")
plt.xlabel("Player Origin")
plt.ylabel("Count")
plt.show()
```

```python
# Top 10 Teams by Players Bought
plt.figure(figsize=(10, 5))
top_teams = df["Team"].value_counts().head(10)
sns.barplot(x=top_teams.values, y=top_teams.index, palette="coolwarm")
plt.title("Top 10 Teams by Players Bought")
plt.xlabel("Number of Players")
plt.ylabel("Team")
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
